```python
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=1000,n_redundant=0,n_features=2,n_clusters_per_class=1,random_state=12,weights=[0.9])
```

```python
X
```

```python
y
```

```python
import pandas as pd
```

```python
df1 = pd.DataFrame(X,columns=['f1','f2'])
df2 = pd.DataFrame(y,columns=['target'])
```

```python
df1.head()
```

```python
df2.head()
```

```python
df_f = pd.concat([df1,df2],axis=1)
df_f.head()
```

```python
df_f['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df_f['f1'],df_f['f2'],c=df_f['target'])
```

```python
from imblearn.over_sampling import SMOTE
```

```python
ove=SMOTE()
X,y = ove.fit_resample(df_f[['f1','f2']],df_f['target'])
```
