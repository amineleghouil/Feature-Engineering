```python
import numpy as np

l = [10,20,30,40,50,60,70,70,70,60,70,80,80,90]
mi,q1,median,q3,max=np.quantile(l,[0,0.25,0.5,0.75,1.0])
```

```python
mi,q1,median,q3,max
```

```python
IQR = (q3-q1)
IQR
```

```python
lower_fence=q1-1.5*(IQR)
higher_fence=q3+1.5*(IQR)
```

```python
lower_fence
```

```python
higher_fence
```
